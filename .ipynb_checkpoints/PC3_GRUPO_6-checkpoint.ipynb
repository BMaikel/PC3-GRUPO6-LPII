## **II.API DEL BCRP**

```python
#La estructura es la siguiente:
"[URL_base]/[códigos de series]/[formato de salida]/[periodo inicial]/[periodo final]/[idioma]"

#Algunos códigos de serie:
"PN01770AM" #PBI
"PN01142MM" #Índide bersatil
"PN01288PM" #Índicede PM sin alimentos
```

Ejemplo de cómo se construye un URL de la API dependiendo del código escogido:

```python
#URL base:
URL_base = "https://estadisticas.bcrp.gob.pe/estadisticas/series/api/"

#Código de series mensuales
cod_ser = "PN01770AM"

#Formato
formato = "/json"

#Periodo
per = "/2005-1/2019-12"

#Idioma
idioma = "/esp"

URL = URL_base + cod_ser + formato + per + idioma

print(URL)
```

## **II. EXTRACCION DE DATOS**

### Extracción del tipo de cambio de las principales modenas desde 1992 hasta 2022

```python
import requests
import pandas as pd
import time

#URL base:
URL_base = "https://estadisticas.bcrp.gob.pe/estadisticas/series/api/"

#Código de series mensuales
cod_ser = ["PN01234PM","PN01235PM","PN01236PM","PN01237PM","PN01238PM","PN01239PM","PN01240PM",
          "PN01241PM","PN01242PM","PN01243PM","PN01244PM"]

#Formato
formato = "/json"

#Periodo
per = "/1992-1/2022-12"

#Idioma
idioma = "/esp"
```

```python
columnas=["Fecha"]
for i in range(len(cod_ser)):
    url = URL_base + cod_ser[i] + formato + per + idioma
    response=requests.get(url)
    jsontemp=response.json()
    columnas.append(jsontemp["config"]["series"][0]["name"])
    if i==0:
        df=pd.json_normalize(jsontemp["periods"])
    else:
        dftemp=pd.json_normalize(jsontemp["periods"])
        df[f'nueva_columna_{i}'] = dftemp["values"]
df=df.set_axis(columnas, axis=1)
```

```python
print(df)
```

#### Exportación de la información an un archivo '.csv'

```python
#Nos quedamos con los valores en lugar de las listas
df.iloc[:,1:] = df.iloc[:,1:].apply(lambda x: x.apply(lambda x: x[0]))

#CSV
df.to_csv('tiposdecambio.csv', header=True)
df
```

### Extracción del PBI

```python
#URL base:
URL_base = "https://estadisticas.bcrp.gob.pe/estadisticas/series/api/"

#Código de series mensuales
cod_ser = 'PN01770AM'

#Formato
formato = "/json"

#Periodo
per = "/2003-1/2022-12"

#Idioma
idioma = "/esp"

URL = URL_base + cod_ser + formato + per + idioma

print(URL)
```

```python
response=requests.get(URL)
jsonform=response.json()
print(jsonform)
```

```python
print(jsonform['config']['series'][0]['name'])
print(jsonform['periods'][0])
```

```python

lista_valores = []
periodos = jsonform['periods']
for i in periodos:
    valores = i['values']
    for v in valores:
        v = float(v)
        lista_valores.append(v)

lista_fechas = []
for i in periodos:
    fechas = i['name']
    lista_fechas.append(fechas)

diccionario= {'Fechas': lista_fechas, 'Producto bruto interno y demanda interna (índice 2007=100) - PBI': lista_valores}
tabla = pd.DataFrame(diccionario)

tabla
```

#### Exportamos a un archivo '.csv'

```python
tabla.to_csv('pbi.csv', header=True)
```

#### Análisis Estadístico

Serie de tiempo para el PBI 

```python
import matplotlib.pyplot as plt
tabla.plot(x='Fechas', y= 'Producto bruto interno y demanda interna (índice 2007=100) - PBI', figsize=(15,5), kind= 'line')
plt.xlabel('Fechas')
plt.ylabel('PBI')
plt.show()
```
